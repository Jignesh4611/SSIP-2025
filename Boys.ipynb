```python
import pandas as pd
import numpy as np

# Load the CSV file
file_path = 'boys_data_.csv'
df = pd.read_csv(file_path)

# Function to generate random weight
import numpy as np

# Function to generate random weight
def generate_random_weight(row):
    category = np.random.choice(
        ['Severe', 'Moderate', 'Normal', 'Above Normal'], 
        p=[0.1, 0.6, 0.2, 0.1]  # Higher chance for "Moderate"
    )
    
    if category == 'Severe':
        return round(np.random.uniform(row['S.W(MAX)'] * 0.8, row['S.W(MAX)']), 2)
    elif category == 'Moderate':
        return round(np.random.uniform(row['M.W(MIN)'], row['M.W(MAX)']), 2)
    elif category == 'Normal':
        return round(np.random.uniform(row['M.W(MAX)'], row['N.W(MAX)']), 2)
    elif category == 'Above Normal':
        return round(np.random.uniform(row['N.W(MAX)'], row['N.W(MAX)'] * 1.2), 2)  # 20% higher than normal

# Function to generate random height
def generate_random_height(row):
    category = np.random.choice(
        ['Severe', 'Moderate', 'Normal', 'Above Normal'], 
        p=[0.1, 0.6, 0.2, 0.1]  # Higher chance for "Moderate"
    )
    
    if category == 'Severe':
        return round(np.random.uniform(row['S.H(MAX)'] * 0.8, row['S.H(MAX)']), 2)
    elif category == 'Moderate':
        return round(np.random.uniform(row['M.H(MIN)'], row['M.H(MAX)']), 2)
    elif category == 'Normal':
        return round(np.random.uniform(row['M.H(MAX)'], row['N.H(MAX)']), 2)
    elif category == 'Above Normal':
        return round(np.random.uniform(row['N.H(MAX)'], row['N.H(MAX)'] * 1.2), 2)  # 20% higher than normal


# Apply the functions to the dataframe
df['Random Weight'] = df.apply(generate_random_weight, axis=1)
df['Random Height'] = df.apply(generate_random_height, axis=1)

# Save updated file
df.to_csv('Updated_Boys_weight_and_height_data.csv', index=False)

# Preview the updated data
print(df.head(10))
```

```python
def classify_weight(row):
    if row['Random Weight'] <= row['S.W(MAX)']:
        return 'S.W'
    elif row['M.W(MIN)'] <= row['Random Weight'] <= row['M.W(MAX)']:
        return 'M.W'
    elif row['Random Weight'] <= row['N.W(MAX)']:
        return 'N.W'
    else:
        return '>N.W'

# Function to classify height category
def classify_height(row):
    if row['Random Height'] <= row['S.H(MAX)']:
        return 'S.H'
    elif row['M.H(MIN)'] <= row['Random Height'] <= row['M.H(MAX)']:
        return 'M.H'
    elif row['M.H(MAX)'] <= row['Random Height'] <= row['N.H(MAX)']:
        return 'N.H'
    else:
        return '>N.H'
```

```python
import pandas as pd
import numpy as np

# Load the reference data
file_path = 'Boys_csv_wasted.csv'  # Reference dataset
bmi = pd.read_csv(file_path)

# Sample random data from mdf


# Function to categorize growth based on height and weight
def categorize_growth(row):
    # Find the row with the closest height from the reference data
    ref_row = bmi.iloc[(bmi['Height(CM)'] - row['Random Height']).abs().idxmin()]
    
    if row['Random Weight'] < ref_row['SAM(MAX)']:
        return 'SAM'
    elif ref_row['MAM(MIN)'] <= row['Random Weight'] <= ref_row['MAM(MAX)']:
        return 'MAM'
    elif ref_row['Normal(MIN)'] <= row['Random Weight'] <= ref_row['Normal(MAX)']:
        return 'Normal'
    elif ref_row['Overweight(MIN)'] <= row['Random Weight'] <= ref_row['Overweight(MAX)']:
        return 'Overweight'
    elif row['Random Weight'] > ref_row['ObeseWeight(MAX)']:
        return 'Obese'
    else:
        return 'Unknown'

# ✅ Apply to entire dataframe
df['Growth Category'] = df.apply(categorize_growth, axis=1)

# ✅ Display the result
print(df[['Random Height', 'Random Weight', 'Growth Category']])
```
